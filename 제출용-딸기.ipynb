# (1) 일조량

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False 
plt.figure(figsize=(3, 5))

aa =pd.read_csv('strawberry.csv',encoding = 'euc-kr')
print(aa.isna().sum())

del aa['지점']
del aa['최저기온(℃)']
del aa['최고기온(℃)']

aa['날짜'] = aa['날짜'].str.replace('\t', '', regex=False)

aa= aa.interpolate(method='linear', limit_direction='both')

print(aa.isna().sum())

aa.boxplot(column=['평균기온(℃)'])
plt.grid(False)
plt.savefig('strawberry_sun_boxplot.png', format='png')
```

```python
aa['날짜'] = pd.to_datetime(aa['날짜'])

aa.set_index(aa['날짜'], inplace=True)

aa_mean = aa['평균기온(℃)'].resample(rule='W-TUE').agg(np.mean).fillna(0)

aa_df = pd.DataFrame({"평균기온(℃)":aa_mean})

aa_3df = aa_df.rolling(window=3).mean().round(1)

aa_sun = aa_3df.dropna()

aa_sun.drop(index=['2024-08-06'])

aa_sun_2022 = aa_sun.loc[:'2022-12-31']
aa_sun_2023 = aa_sun.loc['2023-01-01':]


aa_sun_2022.to_csv("strawberry_sun_2022.csv", index = True)
aa_sun_2023.to_csv("strawberry_sun_2023.csv", index = True)
```

# (2) 강수량

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(3, 5))

bb = pd.read_csv('strawberry_rain.csv',encoding = 'euc-kr')

print(bb.isna().sum())
bb['강수량(mm)'] = bb['강수량(mm)'].fillna(0)
print(bb.isna().sum())

bb.boxplot(column=['강수량(mm)'])
plt.grid(False)
plt.savefig('strawberry_rain_boxplot.png', format='png')
```

```python
del bb['지점']

bb['날짜'] = pd.to_datetime(bb['날짜'])

bb['강수량(mm)'] = bb['강수량(mm)'].fillna(0)

bb.set_index(bb['날짜'], inplace=True)

bb1=bb['강수량(mm)'].resample('W-TUE').agg(np.sum).fillna(0)

bb2 = pd.DataFrame({"강수량(mm)":bb1})

bb_rolling = bb2.rolling(window=3).sum()

bb_rain = bb_rolling.dropna()

bb_rain.drop(index=['2024-08-06'])

bb_rain_2022 = bb_rain.loc[:'2022-12-31']
bb_rain_2023 = bb_rain.loc['2023-01-01':]

bb_rain_2022.to_csv("strawberry_rain_2022.csv", index = True)
bb_rain_2023.to_csv("strawberry_rain_2023.csv", index = True)
```

# 딸기가격 예측모델

# (1) 일조량

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import ensemble
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams as rc


dd =pd.read_csv('strawberry_2022_fin.csv',low_memory=False,encoding='utf-8')
dd_1 = dd[(dd['평균가격'] !=0)]

X = dd_1['평균기온(℃)']
y = dd_1['평균가격']
                                    
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=4200
)        

params = {
    "n_estimators": 2000,
    "max_depth": 4,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "squared_error",
}

X_train = X_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)

reg = ensemble.GradientBoostingRegressor(**params)

reg.fit(X_train, y_train)

a = []
for i in dd_1['평균기온(℃)']:
    a.append(reg.predict([[i]]))
df = pd.DataFrame(a, columns=['예측가격'])

dd_1['DATE'] = dd_1['날짜'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
dd_1 = dd_1.reset_index(drop=True)
dd_3=pd.concat([dd_1,df],axis=1)

fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('thistle')

plt.plot(dd_3['DATE'], dd_3['평균가격'], label='Series 1', marker='o')
plt.plot(dd_3['DATE'] ,dd_3['예측가격'], label='Series 2', marker='x')
plt.savefig('strawberry_sun_plot.png', format='png')
plt.show()
```

```python
dd_3['예측-평균']=dd_3['평균가격']-dd_3['예측가격']
dd_3['예측-평균']=dd_3['예측-평균'].abs().round(2)
dd_3['예측-평균'].mean()
```

# (2) 강수량

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import ensemble
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams as rc

dd =pd.read_csv('strawberry_2022_fin.csv',low_memory=False,encoding='utf-8')
dd_1 = dd[(dd['평균가격'] !=0)]

X = dd_1['강수량(mm)']
y = dd_1['평균가격']
                                  
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=4200
)      

params = {
    "n_estimators": 1600,
    "max_depth": 2,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "squared_error",
}

X_train = X_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)


reg = ensemble.GradientBoostingRegressor(**params)

reg.fit(X_train, y_train)

a = []
for i in dd_1['강수량(mm)']:
    a.append(reg.predict([[i]]))
df = pd.DataFrame(a, columns=['예측가격'])

dd_1['DATE'] = dd_1['날짜'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))

dd_3=pd.concat([dd_1,df],axis=1)

fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('thistle')

plt.plot(dd_3['DATE'], dd_3['평균가격'], label='Series 1', marker='o')
plt.plot(dd_3['DATE'] ,dd_3['예측가격'], label='Series 2', marker='x')
plt.savefig('strawberry_rain_plot.png', format='png')
plt.show()
```

```python
dd_3['예측-평균']=dd_3['평균가격']-dd_3['예측가격']
dd_3['예측-평균']=dd_3['예측-평균'].abs().round(2)
dd_3['예측-평균'].mean()
```
